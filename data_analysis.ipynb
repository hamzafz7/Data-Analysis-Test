```python
# describe: to get statistics info for all features i typically use
print(df['BasePay'].describe())
print(df['OvertimePay'].describe())
print(df['OtherPay'].describe())
print(df['Benefits'].describe())
print(df['TotalPay'].describe())
print(df['TotalPayBenefits'].describe())
```

```python
# handling missing data :
# i noticed for the Notes And Status , all samples are null , so we will drop the two features :
df.drop('Notes', axis=1, inplace=True)
df.drop('Status', axis=1, inplace=True)



```

```python
# now i will deal with basePay Null Value :
'''
 i checked the data and i saw that there is no very clear mode in the data ,
 like i thought if something has not occured too much like 40% percent of times , then
 it is not the bect practice to fill the null values with the mode , therefore i will change it with the median
'''
df['BasePay'].fillna(df['BasePay'].median(), inplace=True)
df['Benefits'].fillna(df['Benefits'].median(),inplace=True)
df['OtherPay'].fillna(df['OtherPay'].median(),inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].median(),inplace=True)
```

```python
# data visulaization :
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['TotalPay'], bins=20, kde=True)
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.title(f'TotalPay Histogram')
plt.show()

sns.histplot(df['TotalPayBenefits'], bins=20, kde=True)
plt.xlabel('TotalPayBenefits')
plt.ylabel('Frequency')
plt.title(f'TotalPayBenefits Histogram')
plt.show()
```

```python
# Count the occurrences of each JobTitle
job_title_counts = df['JobTitle'].value_counts()

# Plotting the pie chart
plt.pie(job_title_counts, labels=job_title_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title('Proportion of Employees in Different Job Titles')
plt.show()
```

```python
columns_of_interest = ['Agency','JobTitle', 'Year']

# Perform grouped analysis for each column using a for loop
for column in columns_of_interest:
    grouped_by_column = df.groupby(column)
    average_salary_by_column = grouped_by_column['TotalPay'].mean(0)
    print(f"Average Salary by {column}:\n", average_salary_by_column)
    print("\n")
```

```python
# let us now do some simple correlation :
correlation_coefficient = df['TotalPay'].corr(df['Benefits'])

print(f"Correlation between Salary and Benefits : {correlation_coefficient}")
```

```python
# Scatter Plot
plt.scatter(df['Benefits'], df['TotalPay'])
plt.xlabel('Benefits')
plt.ylabel('TotalPay')
plt.title(f'Scatter Plot: TotalPay vs Benefits')
plt.show()
```
